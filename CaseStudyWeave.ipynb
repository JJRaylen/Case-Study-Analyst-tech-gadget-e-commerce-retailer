## CLEANING DATA
* Cleaning Empty Cells
* Remove Duplicate
* Cleaning wrong format
* Cleaning wrong data

```python
#CLEANING EMPTY CELLS
df.dropna(inplace = True)
print(df.to_string())
```

```python
#REMOVE DUPLICATE
df.drop_duplicates(inplace = True)
```

```python
df.describe(include="all")
```

```python
# CLEANING WRONG FORMAT
#Convert week's data type from "object" to "datetime"
#df["week"] = pd.to_datetime(df["week"])
```

```python
# QUERY CODE THAT HAVE Weekly_sales == 0
df.loc[df.weekly_sales == 0]
```

```python
#There are 3 cell that wrong data, we can delete or replace them with mean()/median()/Mode()
for x in df.index:
    if df.loc[x,"weekly_sales"] == 0:
        df.drop(x,inplace = True)
```

```python
#CREATE NEW COLLUMN
df["weekly_revenue"] = df["weekly_sales"]*df["price"]
```

# Convert data week -> mm/yy

```python
sample_week="10/31/2016"
month = lambda week:week.split('/')[0]
year = lambda week:week.split('/')[2]
```

```python
df["month"] = df["week"].apply(month)
df["year"] = df["week"].apply(year)
df["date_new"]=df["month"]+'/'+df["year"]
df["date_new"]
```
